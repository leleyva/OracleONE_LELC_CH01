### Importación de datos



```python
#Importo Librerias
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

Adicionamos una columna para identificar la tienda

```python
tienda1['Tienda'] = 'Tienda 1'
tienda2['Tienda'] = 'Tienda 2'
tienda3['Tienda'] = 'Tienda 3'
tienda4['Tienda'] = 'Tienda 4'
tienda1.head()
```

Contactenamos las Tiendas

```python
df = pd.concat([tienda1, tienda2, tienda3, tienda4], ignore_index=True)
```

```python
df.head()
```

#1. Análisis de facturación



```python
#ingresos_por_tienda = df.groupby('Tienda')['Precio'].sum().sort_values(ascending=False)
#print(ingresos_por_tienda)

# Cálculo de ingreso total por tienda
ingresos_por_tienda = df.groupby('Tienda')['Precio'].sum()

# Formatear en pesos colombianos, con separador de miles y 2 decimales
ingresos_formateados = ingresos_por_tienda.apply(lambda x: "$ {:,.2f}".format(x))

print(ingresos_formateados)
```

```python
# Calcular el ingreso total por tienda
ingresos_por_tienda = df.groupby('Tienda')['Precio'].sum().sort_values(ascending=False)

# Crear la gráfica de barras
fig, ax = plt.subplots(figsize=(8,6))
bars = ax.bar(ingresos_por_tienda.index, ingresos_por_tienda.values, color='skyblue')

# Añadir etiquetas con el valor en pesos y centavos
for bar in bars:
    height = bar.get_height()
    ax.annotate(
        f'$ {height:,.2f}',
        xy=(bar.get_x() + bar.get_width() / 2, height),
        xytext=(0, 5),  # desplazamiento vertical sobre la barra
        textcoords="offset points",
        ha='center', va='bottom', fontsize=10, fontweight='bold'
    )

plt.title('Ingresos Totales por Tienda')
plt.ylabel('Ingreso Total (COP)')
plt.xlabel('Tienda')
plt.tight_layout()
plt.show()

```

# 2. Ventas por categoría

```python
# Agrupar por tienda y categoría y contar ventas
ventas_categoria = df.groupby(['Tienda', 'Categoría del Producto']).size()
ventas_categoria = ventas_categoria.reset_index()
ventas_categoria.columns = ['Tienda', 'Categoría del Producto', 'Cantidad Vendida']

# Mostramos la tabla resultante ordenada por tienda y luego por mayor cantidad vendida
ventas_categoria = ventas_categoria.sort_values(['Tienda', 'Cantidad Vendida'], ascending=[True, False])
display(ventas_categoria)
```

```python
# Agrupar y contar la cantidad de ventas por tienda y por categoría
ventas_categoria = df.groupby(['Tienda', 'Categoría del Producto']).size()
ventas_categoria = ventas_categoria.reset_index()
ventas_categoria.columns = ['Tienda', 'Categoría del Producto', 'Cantidad Vendida']

# Ordenar para mostrar las categorías más populares primero en cada tienda
ventas_categoria = ventas_categoria.sort_values(['Tienda', 'Cantidad Vendida'], ascending=[True, False])

# Mostrar los resultados
for idx in range(len(ventas_categoria)):
    print(f"{ventas_categoria.loc[idx, 'Tienda']} - {ventas_categoria.loc[idx, 'Categoría del Producto']}: {ventas_categoria.loc[idx, 'Cantidad Vendida']}")
```

```python
for tienda in df['Tienda'].unique():
    datos = ventas_categoria[ventas_categoria['Tienda'] == tienda]
    plt.figure(figsize=(8, 5))
    plt.barh(datos['Categoría del Producto'], datos['Cantidad Vendida'], color='mediumseagreen')
    plt.title(f'Ventas por Categoría - {tienda}')
    plt.xlabel('Cantidad Vendida')
    plt.ylabel('Categoría del Producto')
    plt.tight_layout()
    plt.show()
```

```python
# Calcula las ventas por categoría para cada tienda
ventas_categoria = df.groupby(['Tienda', 'Categoría del Producto']).size().unstack(fill_value=0)

# Definir posiciones y ancho para las barras
categorias = list(ventas_categoria.columns)
tiendas = list(ventas_categoria.index)
num_tiendas = len(tiendas)
num_categorias = len(categorias)

x = np.arange(num_categorias)
bar_width = 0.18

# Crear gráfico
fig, ax = plt.subplots(figsize=(12, 6))

for i in range(num_tiendas):
    tienda = tiendas[i]
    valores = [ventas_categoria.loc[tienda, categoria] for categoria in categorias]
    ax.bar(x + i*bar_width, valores, width=bar_width, label=tienda)

# Etiquetas y título
ax.set_xticks(x + bar_width*(num_tiendas-1)/2)
ax.set_xticklabels(categorias, rotation=35, ha='right')
ax.set_ylabel('Cantidad Vendida')
ax.set_xlabel('Categoría del Producto')
ax.set_title('Ventas por Categoría en Cada Tienda')
ax.legend(title='Tienda')

plt.tight_layout()
plt.show()
```

```python
# Agrupar y contar la cantidad de ventas por tienda y por categoría
ventas_categoria = df.groupby(['Tienda', 'Categoría del Producto']).size()
ventas_categoria = ventas_categoria.reset_index()
ventas_categoria.columns = ['Tienda', 'Categoría del Producto', 'Cantidad Vendida']

# Mostrar un gráfico circular por cada tienda
tiendas = ventas_categoria['Tienda'].unique()

for tienda in tiendas:
    data_tienda = ventas_categoria[ventas_categoria['Tienda'] == tienda]
    plt.figure(figsize=(7, 7))
    plt.pie(
        data_tienda['Cantidad Vendida'],
        labels=data_tienda['Categoría del Producto'],
        autopct='%1.1f%%',
        startangle=140,
        textprops={'fontsize': 10}
    )
    plt.title(f'Participación de Categorías en Ventas - {tienda}', fontsize=14, fontweight='bold')
    plt.axis('equal')  # Hace que el gráfico sea un círculo perfecto
    plt.tight_layout()
    plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calcular la valoración promedio por tienda
valoracion_media = df.groupby('Tienda')['Calificación'].mean().sort_values(ascending=False)

# Presentar los resultados con 2 decimales, sin usar lambda
valoracion_media_formateada = {}
for tienda in valoracion_media.index:
    promedio = valoracion_media[tienda]
    promedio_formateado = "{:.2f}".format(promedio)
    valoracion_media_formateada[tienda] = promedio_formateado

# Imprimir el resultado
for tienda, promedio in valoracion_media_formateada.items():
    print(f"{tienda}: {promedio}")
```

```python
import matplotlib.pyplot as plt

# Calcular la valoración promedio por tienda
valoracion_media = df.groupby('Tienda')['Calificación'].mean().sort_values(ascending=False)

# Crear el gráfico de barras
fig, ax = plt.subplots(figsize=(8, 6))
bars = ax.bar(valoracion_media.index, valoracion_media.values, color='mediumslateblue')

# Añadir etiquetas de los valores sobre cada barra
for bar in bars:
    promedio = bar.get_height()
    etiqueta = "{:.2f}".format(promedio)
    ax.annotate(
        etiqueta,
        xy=(bar.get_x() + bar.get_width() / 2, promedio),
        xytext=(0, 5),  # desplazamiento hacia arriba
        textcoords='offset points',
        ha='center', va='bottom', fontsize=11, fontweight='bold'
    )

plt.title('Valoración Media por Tienda', fontsize=16, fontweight='bold')
plt.ylabel('Calificación Promedio', fontsize=12)
plt.xlabel('Tienda', fontsize=12)
plt.ylim(0, 5)  # Limitar de 0 a 5 para mayor claridad
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
# 1. Contar ventas por producto en cada tienda
ventas_producto = df.groupby(['Tienda', 'Producto']).size().reset_index(name='Cantidad Vendida')

# 2. Identificar el más vendido y el menos vendido en cada tienda
tabla_resumen = []
tiendas = ventas_producto['Tienda'].unique()

for tienda in tiendas:
    datos_tienda = ventas_producto[ventas_producto['Tienda'] == tienda]
    max_ventas = datos_tienda['Cantidad Vendida'].max()
    min_ventas = datos_tienda['Cantidad Vendida'].min()
    productos_max = datos_tienda[datos_tienda['Cantidad Vendida'] == max_ventas]
    productos_min = datos_tienda[datos_tienda['Cantidad Vendida'] == min_ventas]
    # Es posible que haya empate, por eso podría haber más de un producto más/menos vendido

    for idx, fila in productos_max.iterrows():
        tabla_resumen.append({
            'Tienda': tienda,
            'Producto': fila['Producto'],
            'Tipo': 'Más vendido',
            'Cantidad Vendida': fila['Cantidad Vendida']
        })
    for idx, fila in productos_min.iterrows():
        tabla_resumen.append({
            'Tienda': tienda,
            'Producto': fila['Producto'],
            'Tipo': 'Menos vendido',
            'Cantidad Vendida': fila['Cantidad Vendida']
        })

# 3. Convertir a DataFrame para visualización clara
import pandas as pd
tabla_resumen_df = pd.DataFrame(tabla_resumen)

# 4. Mostrar la tabla ordenada
tabla_resumen_df = tabla_resumen_df.sort_values(['Tienda', 'Tipo'], ascending=[True, True])
display(tabla_resumen_df)

```

```python


# Contar ventas por producto en cada tienda
ventas_producto = df.groupby(['Tienda', 'Producto']).size().reset_index(name='Cantidad Vendida')

# Identificar más y menos vendidos por tienda
tabla_resumen = []
tiendas = ventas_producto['Tienda'].unique()

for tienda in tiendas:
    datos_tienda = ventas_producto[ventas_producto['Tienda'] == tienda]
    max_ventas = datos_tienda['Cantidad Vendida'].max()
    min_ventas = datos_tienda['Cantidad Vendida'].min()
    productos_max = datos_tienda[datos_tienda['Cantidad Vendida'] == max_ventas]
    productos_min = datos_tienda[datos_tienda['Cantidad Vendida'] == min_ventas]
    # Puedes tener empates
    for _, fila in productos_max.iterrows():
        tabla_resumen.append({
            'Tienda': tienda,
            'Producto': fila['Producto'],
            'Tipo': 'Más vendido',
            'Cantidad Vendida': fila['Cantidad Vendida']
        })
    for _, fila in productos_min.iterrows():
        tabla_resumen.append({
            'Tienda': tienda,
            'Producto': fila['Producto'],
            'Tipo': 'Menos vendido',
            'Cantidad Vendida': fila['Cantidad Vendida']
        })

tabla_resumen_df = pd.DataFrame(tabla_resumen)

# Visualización: gráfico de barras horizontal agrupado
fig, ax = plt.subplots(figsize=(12, 7))

# Asignar un color distinto para "Más vendido" y "Menos vendido"
colores = {'Más vendido': 'mediumseagreen', 'Menos vendido': 'tomato'}

# Para mejor visual, juntar nombre producto y tienda
tabla_resumen_df['Etiqueta'] = tabla_resumen_df['Producto'] + ' (' + tabla_resumen_df['Tienda'] + ')'

for tipo in ['Más vendido', 'Menos vendido']:
    datos = tabla_resumen_df[tabla_resumen_df['Tipo'] == tipo]
    ax.barh(datos['Etiqueta'], datos['Cantidad Vendida'], color=colores[tipo], label=tipo)

# Mostrar etiquetas de valor
for i, fila in tabla_resumen_df.iterrows():
    ax.text(
        fila['Cantidad Vendida'] + 0.05,
        i,
        str(fila['Cantidad Vendida']),
        va='center',
        fontsize=10,
        fontweight='bold'
    )

ax.set_xlabel('Cantidad Vendida')
ax.set_title('Productos Más y Menos Vendidos por Tienda')
ax.legend(title='Tipo de producto')
plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio por tienda
envio_promedio = df.groupby('Tienda')['Costo de envío'].mean().sort_values(ascending=False)

# Formatear con separador de miles y dos decimales (sin lambda)
envio_promedio_formateado = {}
for tienda in envio_promedio.index:
    valor = envio_promedio[tienda]
    valor_formato = "$ {:,.2f}".format(valor)
    envio_promedio_formateado[tienda] = valor_formato

# Imprimir resultado claro
print("Costo de envío promedio por tienda:")
for tienda, valor in envio_promedio_formateado.items():
    print(f"{tienda}: {valor}")

```
